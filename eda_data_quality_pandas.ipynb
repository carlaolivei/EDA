<h3>Dados numéricos ausentes: substituição por média</h3>

```python
#Calculando a média da coluna "normalized-losses"
media_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Média de normalized-losses:", media_norm_loss)
#Substituindo "NaN" pelo valor da média na coluna "normalized-losses"
df["normalized-losses"].replace(np.nan, media_norm_loss, inplace=True)
```

```python
#Calculando a média para a coluna "bore"
media_bore=df['bore'].astype('float').mean(axis=0)
print("Média bore:", media_bore)
#Substituindo NaN pelo valor da média na coluna "bore"
df["bore"].replace(np.nan, media_bore, inplace=True)
```

```python
#Calculando a média da coluna "stroke"
media_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Média stroke:", media_stroke)
#Substituindo NaN pela média na coluna "stroke"
df["stroke"].replace(np.nan, media_stroke, inplace = True)
```

```python
#Calculando a média para a coluna "horsepower"
media_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Média horsepower:", media_horsepower)
#Substituindo "NaN" pelo valor da média na coluna "horsepower"
df['horsepower'].replace(np.nan, media_horsepower, inplace=True)
```

```python
#Calculando a média para a coluna "peak-rpm"
media_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Média peak rpm:", media_peakrpm)
#Substituindo NaN pelo valor do media na coluna "peak-rpm"
df['peak-rpm'].replace(np.nan, media_peakrpm, inplace=True)
```

<h3>Dados categóricos ausentes: substituição por moda (dados mais frequentes)</h3>

Neste caso devemos verificar o valor mais frequente para substituir os dados ausentes.

```python
#Utilização da função describe(include = 'all') para verificar os dados com maior frequência para coluna 'num-of-doors'
df['num-of-doors'].describe(include = 'all')
```

```python
#Para valores mais frequentes em uma coluna específica, podemos usar o método ".value_counts ()":
df['num-of-doors'].value_counts()
```

```python
#Também podemos usar o método ".idxmax ()" para calcular automaticamente o tipo mais comum
df['num-of-doors'].value_counts().idxmax()
```

O procedimento de substituição é muito semelhante ao que vimos anteriormente

```python
#substituindo os valores de 'num-of-doors' ausentes pelos mais frequentes
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

<h3>Finalmente, vamos eliminar todas as linhas que não têm dados na coluna price</h3>

```python
# remoção das linhas com NaN na coluna "price"
df.dropna(subset=["price"], axis=0, inplace=True)
# redefinindo o índice devido remoção das linhas
df.reset_index(drop=True, inplace=True)
```

<h2> Padronização de dados (transformação de atributos numéricos) </h2>

A padronização de dados consiste em um processo de transformar dados em um formato comum que permite ao analista de dados fazer uma comparação significativa.

<b> Exemplo </b>
<p> Transformação de mpg em L / 100km: </p>
<p> Em nosso conjunto de dados, as colunas de consumo de combustível "city-mpg" e "highway-mpg" são representadas pela unidade de mpg (milhas por galão). Suponha que estejamos desenvolvendo um aplicativo em um país que aceita o consumo de combustível com o padrão L / 100km </p>
<p> Precisamos aplicar <b> transformação de dados </b> para transformar mpg em L / 100km. </p>

A fórmula para esta conversão de unidades é: L / 100km = 235 / mpg.

```python
# Conversão de mpg em L / 100km utilizando operação matemática (235 dividido por mpg)
df['city-L/100km'] = 235/df["city-mpg"]
# verificando o dataframe após a conversão
df.head()
```

```python
# Conversão de mpg em L / 100km utilizando operação matemática (235 dividido por mpg)
df['highway-L/100km'] = 235/df["highway-mpg"]
# verificando o dataframe após a conversão
df.head()
```

<h2> Normalização de dados (transformação de atributos numéricos) </h2>

<p> Normalização é o processo de transformar valores de várias variáveis em um intervalo semelhante. As normalizações típicas incluem dimensionar a variável para que a média da variável seja 0, dimensionar a variável para que a variância seja 1 ou dimensionar a variável para que os valores das variáveis variem de 0 a 1. No scikit-lear, por exemplo, existems muitos meétodos para normalização de dados.
</p>

<b> Exemplo </b>
<p> Para demonstrar a normalização, digamos que queremos dimensionar as colunas "length", "width" e "height" </p>
<p> <b> Destino: </b> normalizar essas variáveis para que seus valores variem de 0 a 1. </p>
<p> <b> Abordagem: </b> substituir o valor original por (valor original) / (valor máximo) </p>

```python
# substituindo (valor original) por (valor original) / (valor máximo)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
# mostrar as colunas após a normalização
df[["length","width","height"]].head()
```

<h2> Binning (conversão númerico-categórica) </h2>

<p>
     Binning é um processo de transformação de variáveis numéricas contínuas em 'bins' categóricos, para análise agrupada.
</p>

<b> Exemplo: </b>
<p> Em nosso conjunto de dados, "horsepower" é uma variável numérica variando de 48 a 288, ela tem 57 valores únicos. E se nos preocupássemos apenas com a diferença de preço entre carros com alta, média e pequena potência (3 tipos)? Podemos reorganizá-los em três 'faixas' para simplificar a análise. </p>

<p> Usaremos o método Pandas 'cut' para segmentar a coluna de 'horsepower' em 3 tipos. </p>

<h3> Exemplo de binning (agrupamento) de dados com Pandas </h3>

```python
#Convertendo os dados para o formato correto
df["horsepower"]=df["horsepower"].astype(int, copy=True)
df["horsepower"].dtype
```

Vamos plotar o histograma da coluna "horsepower", para ver como é a distribuição da potência do motor

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"],color='#00BFFF')

# definição dos rótulos x / y e do título do gráfico
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title(" Histograma da coluna horsepower")
```

NOTA: o exemplo mostra um histograma com assimetria à direita.

```python
#Construção de uma matriz bin
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#Definição dos nomes das faixas
group_names = ['Baixo', 'Médio', 'Alto']
```

```python
#Aplicação da função "cut" para determinar a que cada valor de "df ['horsepower']" pertence.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Verificação do número de automóveis em cada faixa
df["horsepower-binned"].value_counts()
```

Agora vamos gerar um grafico de barra para verificar a distribuição de cada grupo

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

barra_cor = 'green'
pyplot.bar(group_names, df["horsepower-binned"].value_counts(),color=barra_cor)

# definindo os rótulos x / y e título do gráfico
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("Gráfico de barra de coluna horsepower")
```

O gráfico de barra acima mostra os bins (categorias "Baixo", "Médio" e "Alto") que foram criados para "horsepower".

IMPORTANTE: foi reduzido os intervalos de 57 para 3.

Vamos utilizar um histograma para visualizar a distribuição dos bins que foram criados.

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

# desenhando o histograma do atributo "horsepower" com grupos = 3
plt.pyplot.hist(df["horsepower"], bins = 3, color='r')

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("Histograma da coluna horsepower com os bins")
```

O histograma acima mostra o resultado do "binning" para o atributo "horsepower" e que o formato do histograma se manteve o mesmo antes dessa transformação, ou seja, continuou com assimitria à direita.

<h2> Variável indicadora ou dummy (conversão simbólico-numérico) </h2>

<p>
     Uma variável indicadora (ou variável dummy) é uma variável numérica usada para rotular categorias. Eles são chamados de 'manequins' porque os próprios números não têm um significado inerente.
</p>

<b> Por que utilizar variáveis indicadoras? </b>
<p>
     Para poder utilizar variáveis categóricas para análise de regressão linear, por exemplo.
</p>
<b> Exemplo:</b>
<p>
     A coluna "fuel-type" tem dois valores exclusivos, "gas" ou "diesel". A regressão não compreende palavras, apenas números. Para usar este atributo na análise de regressão, é necessário converter "fuel-type" em variáveis indicadoras.
</p>

<p>
     Para isso será utilizado o método 'get_dummies', do Pandas, para atribuir valores numéricos a diferentes categorias de tipo de combustível ("fuel-type").
</p>

```python
#Vamos obter a variável indicadora e atribuí-la ao dataframe "dummy_variable_1"
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#Após obter a variável indicadora e atribuí-la é necessário alterar os nomes das colunas para maior clareza
dummy_variable_1.rename(columns={'fuel-type-gas':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

Agora temos os valores 0 e 1 para representar "gas" e "diesel" na coluna "fuel-type". Quando o valor for "gas" será preenchido com 1 e "diesel" com 0 e vice-versa.

Para concluir este pré-processamento de dados deve-se inserir esta coluna no conjunto de dados original.

```python
#Inclusão de "dummy_variable_1" no dataframe "df"
df = pd.concat([df, dummy_variable_1], axis=1)

# Remoção da coluna original "fuel-type" de "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

As duas últimas colunas são agora a representação da variável indicadora da variável do tipo de combustível. Agora são todos 0s e 1s.
